```python
!pip install matplotlib
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np  # For random number generation in Step 4

# 1: Create a simple line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# a. Plot the data
plt.plot(x, y)

# b. Customize the plot by adding a title, axis labels, and a grid
plt.title('Simple Line Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)

# Display the plot
plt.show()

# 2: Create a bar graph
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

# a. Plot the data as a bar graph
plt.bar(students, marks)

# b. Customize the colors and add a title
plt.bar(students, marks, color=['blue', 'green', 'red', 'orange'])
plt.title('Student Marks in Subject')

# Display the plot
plt.show()

# 3: Create a pie chart
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# a. Create a pie chart with the region names as labels
plt.pie(revenue, labels=regions)

# b. Highlight the region with the highest revenue (North America)
explode = [0.1, 0, 0, 0]  # Explode the first slice (North America)
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%')
plt.title('Revenue Distribution by Region')

# Display the plot
plt.show()

# 4: Generate a histogram
# Generate 1000 random integers between 1 and 100
data = np.random.randint(1, 101, size=1000)

# Create the histogram
plt.hist(data, bins=20, edgecolor='black')
plt.title('Frequency Distribution of Random Integers (1-100)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
