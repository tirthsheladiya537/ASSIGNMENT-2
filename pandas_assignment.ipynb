```python
!pip install pandas
# Cell 1: Import Pandas
import pandas as pd

# Cell 2: Create the DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)

# a. Print the first five rows of the DataFrame
print("First five rows of the DataFrame:")
print(df.head())
print("\n")

# b. Get the summary statistics of the 'Age' and 'Salary' columns
print("Summary statistics for 'Age' and 'Salary':")
print(df[['Age', 'Salary']].describe())
print("\n")

# c. Calculate the average salary of employees in the 'HR' department
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print(f"Average salary in HR department: {hr_avg_salary}")
print("\n")

# Cell 3: Add a new column 'Bonus' which is 10% of the salary
df['Bonus'] = df['Salary'] * 0.1
print("DataFrame after adding 'Bonus' column:")
print(df)
print("\n")

# Cell 4: Filter the DataFrame to show employees aged between 25 and 30
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print("Employees aged between 25 and 30:")
print(filtered_df)
print("\n")

# Cell 5: Group the data by 'Department' and calculate the average salary for each department
dept_avg_salary = df.groupby('Department')['Salary'].mean()
print("Average salary by department:")
print(dept_avg_salary)
print("\n")

# Cell 6: Sort the DataFrame by 'Salary' in ascending order and save to a new CSV file
df_sorted = df.sort_values('Salary')
print("DataFrame sorted by Salary (ascending):")
print(df_sorted)
df_sorted.to_csv('sorted_salaries.csv', index=False)
print("Sorted DataFrame saved to 'sorted_salaries.csv'")
```
